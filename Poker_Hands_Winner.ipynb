```python
import math
import os
import random
import re
import sys


def handwin(hand):
    values0 = [c[0] for c in hand]
    suits0 = [c[1] for c in hand]
    
    values = []
    for i in values0:
        if i == 'T': values.append(10)
        elif i == 'J': values.append(11)
        elif i == 'Q': values.append(12)
        elif i == 'K': values.append(13)
        elif i == 'A': values.append(14)
        else: values.append(int(i))
    
    hand1v = sorted(values[:5])
    hand2v = sorted(values[5:])
    suit1 = suits0[:5]
    suit2 = suits0[5:]
    
    hand1score = []
    hand2score = []
    
    def score(hand,suit):
        #10 - royal flush
        #9 - straight flush
        #8 - four of a kind
        #7- full house
        #6 - flush
        #5 - straigh
        #4 - three of a kind
        #3 - two pair
        #2 - pair
        #1 - high card
        hands = sorted(hand)
        same_suit = all(c == suit[0] for c in suit)
        in_seq = (hands in (list(range(hands[0],hands[4]+1,1)), [2,3,4,5,14]))
        
        if hands == ([10,11,12,13,14] and same_suit): return 10, None
        if (in_seq and same_suit): return 9, min(hands)
        
        length = len(set(hands))
        if length == 2: 
            if hands.count(hands[2]) == 4: return 8, ([hands.count(i) == 4 for i in hands].index(True), [hands.count(i) == 1 for i in hands].index(True)) # output the index of the quadraple and single member
            elif hands.count(hands[2]) == 3: return 7, ([hands.count(i) == 3 for i in hands].index(True), [hands.count(i) == 2 for i in hands].index(True)) # output the index of the triple and double member
        
        if (same_suit and (not in_seq)):
            return 6, hands
        if (in_seq and (not same_suit)):
            return 5, min(hands)
        if length == 3:
            if hands.count(hands[2]) == 3: return 4, hands #output the index of the triple first single ones 
            else: return 3, hands
        if length == 4:
            return 2, hands
        if ((not in_seq) and (not same_suit)): return 1, hands
    
    hand1score = score(hand1v, suit1)
    hand2score = score(hand2v, suit2)
    
    h1score = hand1score[0]
    h1para = hand1score[1]
    h2score = hand2score[0]
    h2para = hand2score[1]
    
    if h1score > h2score: 
        return "Player 1"
    elif h1score < h2score:
        return "Player 2"
    elif h1score == h2score:
        if h1score in (9,5): #straight flush or straight
            if h1para > h2para:
                return "Player 1"
            else: return "Player 2"
        if h1score in (8,7): # four of a kind and full house
            if hand1v[h1para[0]] > hand2v[h2para[0]]:
                return "Player 1"
            elif hand1v[h1para[0]] < hand2v[h2para[0]]:
                return "Player 2"
            elif hand1v[h1para[0]] == hand2v[h2para[0]]:
                if hand1v[h1para[1]] > hand2v[h2para[1]]:
                    return "Player 1"
                else: return "Player 2"
        if h1score in (6,1): #flush or high card
            h1 = sorted(h1para, reverse = True)
            h2 = sorted(h2para, reverse = True)
            if h1 > h2: 
                return "Player 1"
            else: return "Player 2"
        if h1score == 4: #three of a kind
            if hand1v[2] > hand2v[2]: 
                return "Player 1"
            elif hand1v[2] < hand2v[2]:
                return "Player 2"
            elif hand1v[2] == hand2v[2]:
                a1 = [h1para.count(i) for i in h1para]
                n1 = [i for i in range(len(a1)) if a1[i] == 1]
                a2 = [h2para.count(i) for i in h2para]
                n2 = [i for i in range(len(a2)) if a2[i] == 1]
                m1 = max([hand1v[i] for i in n1])
                m2 = max([hand2v[i] for i in n2])
                if m1 > m2: 
                    return "Player 1"
                else: return "Player 2"
        if h1score == 3: #two pair
            pair1 = [i for i in h1para if h1para.count(i) == 2]
            pair2 = [i for i in h2para if h2para.count(i) == 2]
            if max(pair1) > max(pair2):
                return "Player 1"
            elif max(pair1) < max(pair2):
                return "Player 2"
            elif max(pair1) == max(pair2):
                if min(pair1) > min(pair2):
                    return "Player 1"
                elif min(pair1) < min(pair2):
                    return "Player 2"
                elif pair1 == pair2:
                    s1 = [i for i in h1para if h1para.count(i) == 1]
                    s2 = [i for i in h2para if h2para.count(i) == 1]
                    if s1 > s2: 
                        return "Player 1"
                    else: return "Player 2"
        if h1score == 2: #pair
            pair1 = [i for i in h1para if h1para.count(i) == 2]
            pair2 = [i for i in h2para if h2para.count(i) == 2]
            s1 = [i for i in h1para if h1para.count(i) == 1]
            s2 = [i for i in h2para if h2para.count(i) == 1]
            if pair1 > pair2:
                return "Player 1"
            elif pair1 < pair2:
                return "Player 2"
            elif pair1 == pair2:
                s11 = sorted(s1, reverse = True)
                s22 = sorted(s2, reverse = True)
                if s11 > s22:
                    return "Player 1"
                else: return "Player 2"

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')
    n = input()
    num = int(n)
    queries = []

    for _ in range(num):
        queries.append(list(map(str, input().rstrip().split())))
    results = []
    for i in range(num):
        results.append(handwin(queries[i]))
    
    for i in range(num):
        fptr.write(str(results[i]) + '\n')

    fptr.close()
```
